```python
import requests # library to handle requests
from bs4 import BeautifulSoup # library to parse HTML and XML documents
import pandas as pd # library for data analsysis
```

```python
url = 'https://www.census2011.co.in/district.php'
```

```python
source = requests.get(url).text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
table = str(soup.table)
```

```python
from IPython.display import display_html
display_html(table ,raw = True)
```

```python

df = pd.read_html(table)
df
```

```python

df_data = df[0]
df_data
```

```python

df_data.rename(columns = {'Sex-Ratio':'SexRatio'}, inplace = True)
```

```python
df_data.shape
```

```python
df_data['Growth'] = df_data['Growth'].map(lambda x: x.rstrip('%'))
df_data
```

```python
df_data['Growth'] = df_data['Growth'].astype(float)
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
avg_Literacy = df_data['Literacy'].mean()
avg_Literacy
```

```python
df_literacy72 = df_data[df_data.Literacy <= 72.30]
df_literacy72
```

```python
plt.scatter(df_literacy72.Population, df_literacy72.Literacy)
plt.xlabel("Population") 
plt.ylabel("Literacy Rate") 
plt.title("Populaion V/s Literacy Rate") 
plt.show()
```

```python

df_literacy50 = df_data[df_data.Literacy <= 50.00]
df_literacy50
```

```python

plt.plot(df_literacy50.Literacy, df_literacy50.District, color = 'blue')
plt.plot(df_literacy50.Growth, df_literacy50.District, color = 'red')
plt.ylabel("District") 
plt.xlabel("Literacy Rate and Growth Rate") 
plt.title("Trends in Literacy Rates and Growth Rates") 
plt.show()
```

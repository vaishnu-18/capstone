# Segmenting and Clustering Toronto 3

```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
```

```python
source = requests.get(url).text
```

```python
soup = BeautifulSoup(source, 'lxml')
```

```python
table = str(soup.table)
```

```python
from IPython.display import display_html
display_html(table ,raw = True)
```

```python
df = pd.read_html(table)
df_new = df[0]
df_new
```

```python
df_new = df_new[df_new.Borough != 'Not assigned']
df_new
```

```python
df_new = df_new.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df_new.reset_index(inplace=True)
df_new
```

```python
import numpy as np 
df_new['Neighborhood'] = np.where(df_new['Neighborhood'] == 'Not assigned',df_new['Borough'], df_new['Neighborhood'])
df_new
```

```python
df_new.shape
```

```python
df_lat_long = pd.read_csv('https://cocl.us/Geospatial_data')
df_lat_long
```

```python
df_merge = pd.merge(df_new,df_lat_long,on='Postal Code')
df_merge
```

```python
df_cluster = df_merge[df_merge['Borough'].str.contains('Toronto',regex=False)]
df_cluster
```

```python
from sklearn.cluster import KMeans
import matplotlib.cm as cm
import matplotlib.colors as colors

k=5
df_toronto = df_cluster.drop(['Postal Code','Borough','Neighbourhood'],1)
kmeans = KMeans(n_clusters = k,random_state=0).fit(df_toronto)
kmeans.labels_
df_cluster.insert(0, 'Cluster Labels', kmeans.labels_)
df_cluster
```
